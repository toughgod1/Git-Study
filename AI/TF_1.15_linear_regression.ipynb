```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
tf.enable_eager_execution()
tf.__version__
```

```python
X = np.array([2, 4, 6, 8], dtype="float32")
Y = np.array([81, 93, 91, 97], dtype="float32")
```

```python
W = tf.Variable([2], dtype="float32")
b = tf.Variable([1], dtype="float32")
```

```python
W
```

```python
b
```

```python
learning_rate = 0.01
```

```python
for i in range(1000):
    with tf.GradientTape() as tape:
        hypothesis = W * X + b
        print("hypothesis:", hypothesis)
        cost = tf.reduce_mean(tf.square(hypothesis - Y))
        print("cost", cost)
        W_grad, b_grad = tape.gradient(cost, [W, b])
        print("W_grad:", W_grad, ":b_grad:", b_grad)
        W.assign_sub(W_grad*learning_rate)
        b.assign_sub(b_grad*learning_rate)
```

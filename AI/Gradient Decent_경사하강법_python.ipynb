```python
import numpy as np
```

```python
x=np.array([1,2,3], dtype="float32")
y=np.array([2, 2.5, 3.5], dtype="float32")
```

```python
x
```

```python
y
```

```python
w = 2
b = 1
```

```python
hypothesis = x*w + b
hypothesis
```

```python
w_gred = 1/3 *np.dot((hypothesis-y), x.T)
w_gred
```

```python
w = w - w_gred
w
```

```python
b_gred= 1/3 * np.sum(hypothesis-y)
b_gred
```

```python
b = b - b_gred
b
```

```python
for i in range(100):
    hypothesis = x*w + b
    MSE = np.sum((hypothesis-y)**2)
    print("MSE:", MSE)
    w_gred = 1/3*np.dot((hypothesis-y), x.T)
    print("w_gred: ", w_gred)
    b_gred = 1/3 * np.sum(hypothesis-y)
    print("b_gred:", b_gred)
    w=w-w_gred
    print("w:",w)
    b=b-b_gred
    print("b:", b)
```
